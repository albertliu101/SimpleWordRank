```python
UniqueWordsBySite = []
AllUniqueWords = {}
for i in range(len(webDF)):
    currDict = {}
    currH1 = webDF.iloc[i]['h1']
    currH2 = webDF.iloc[i]['h2']
    currDiv = webDF.iloc[i]['div']
    # Lets loop through h1
    badCharacters = ['\xa0','\n','\t',':',';', '?','.',',','(',')','\\','/']
    webList = []
    for el in currDiv:
        elTrim = el
        for bad in badCharacters:
            elTrim = elTrim.replace(bad,' ')
        elList = elTrim.split(' ')
        for word in elList:
            if word != '':
                webList.append(word)
    for word in webList:
        if word.lower() not in currDict.keys():
            currDict[word.lower()] = 1
        else:
            currDict[word.lower()] += 1
        if word.lower() not in AllUniqueWords.keys():
            AllUniqueWords[word.lower()] = 1
        else:
            AllUniqueWords[word.lower()] += 1
    UniqueWordsBySite.append(currDict)
            
        
```

```python
# Sort all Unique words by frequency
import operator
keys = AllUniqueWords.keys()
sorted_tuples = sorted(AllUniqueWords.items(),key=operator.itemgetter(1),reverse=True)
sorted_UniqueWords = {k:v for k,v in sorted_tuples}
```
